```python
from sentence_transformers import SentenceTransformer

# 모델 로드
model = SentenceTransformer('jhgan/ko-sbert-sts')

# 입력 문장
sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."]

# 문장 임베딩 생성
embeddings = model.encode(sentences)

print("Embeddings:", embeddings)
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('jhgan/ko-sbert-sts')

# 입력 문장
sentence1 = "나는 사과를 좋아합니다."
sentence2 = "너는 사과를 좋아하니?"

# 임베딩 생성
embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)

# 코사인 유사도 계산
cosine_similarity = util.cos_sim(embedding1, embedding2)
print("Cosine Similarity:", cosine_similarity)
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('jhgan/ko-sbert-multitask') # 모델을 바꿔 봄

# 입력 문장
sentence1 = "저는 밝고 긍정적인 에너지를 가진 사람입니다! 날씨가 좋을 때는 노래를 들으면서 혼자 산책하는 것을 좋아합니다. 특히 힙합이나 최신 아이돌 노래를 즐겨듣는 편입니다!! 저는 현재 주어진 일에 최선을 다하는 성실한 성격이에요! 거짓없고 솔직하고 털털한 편입니다! 맛있는 음식을 먹으러 다니는 것을 좋아하고 술 마시는 것도 좋아합니다! 제일 좋아하는 음식은 육회, 국밥, 마라탕입니다"
sentence2 = "저는 기력이 별로 없는 사람입니다. 스트레스와 피로가 쌓일 때는 반려견과 함께 누워있음으로써 심신의 안정을 취합니다. 강아지 중 특히 단모 치와와를 좋아합니다. 저는 연락을 자주하지 않는 것을 선호합니다. 자주 전화하는것을 별로 좋아하지 않습니다. 술 마시는 것을 별로 좋아하지 않습니다. 기력이 별로 없기 때문입니다. 좋아하는 음식은 샤브샤브, 초밥, 국밥입니다."

# 임베딩 생성
embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)

# 코사인 유사도 계산
cosine_similarity = util.cos_sim(embedding1, embedding2)
print("Cosine Similarity:", cosine_similarity)
```
